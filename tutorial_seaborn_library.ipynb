### Import seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Sample dataset

```python
print(sns.get_dataset_names())
```

```python
titanic_df = sns.load_dataset('titanic')

titanic_df.head()
```

### Distribution Plot

```python
sns.displot(titanic_df['age'], kde=True)
```

### Joint Plot

```python
sns.jointplot(x='age', y='fare', data=titanic_df, kind='reg')
plt.show()
```

### Kernel Distribution Estimation(KDE) Plot

```python
sns.kdeplot(x=titanic_df['age'], y=titanic_df['fare'], shade=True)
plt.show()
```

### Pair Plot

```python
selected_columns = ['survived','pclass','age', 'fare', 'sex']
df = titanic_df[selected_columns]

sns.pairplot(df, hue='sex')
plt.show()
```

### Bar Plot

```python
titanic_df.head()
```

```python
sns.barplot(x='alive',y='age',data=titanic_df)
plt.show()
```

### Count Plot

```python
sns.countplot(x='alive', data=titanic_df)
plt.show()
```

### Box Plot

```python
sns.boxplot(x='class', y='fare', data=titanic_df, hue='sex')
plt.show()
```

### Scatter Plot

```python
sns.scatterplot(x='age', y='fare', data=titanic_df)
plt.show()
```

### Heatmap 

```python
iris_df = sns.load_dataset('iris')
corr_matrix = iris_df.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```
