```python
import matplotlib.pyplot as plt
```

### Line plot

```python
x = [2,3,4,5]
y = [4,6,8,10]

plt.plot(x,y)

plt.show()
```

```python
x = [2,8,9,12]
y = [3,4,7,3]

plt.plot(x,y,color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12 )
plt.title('Sample graph', fontdict={'family':'fantasy','color':'blue','size':20})
plt.xlabel('Luck')
plt.ylabel('Success')


plt.show()
```

### Bar chart

```python
languages = ['Java', 'C', 'PHP', 'Python']
popularity = [80, 60, 50, 90]
plt.title('Popularity of Computer Languages', fontdict={'family':'cursive','color':'blue','size':20})

plt.bar(languages, popularity, color='green')
plt.xlabel('Languages')
plt.ylabel('Popularity')

plt.show()
```

### Pie chart

```python
Languages = ['English', 'Mandarin', 'Hindi', 'Spanish', 'French', 'Modern Standard Arabic', 'Russian']
Population = [1452, 1118, 602, 548, 280, 274, 258]

plt.pie(Population, labels=Languages, autopct='%1.2f%%')

plt.title(label="Most Spoken Languages in World",  
          fontstyle='italic') 

plt.show()
```

### Histogram

```python
import numpy as np

x = np.random.normal(200, 5, 500) 

plt.hist(x, bins=30)

plt.hist(x)
plt.show()
```

### Scatter plot

```python
air_temparature = [17,22,23,25,28,30,35,38]
ac_sales = [1,5,7,9,12,22,50,45]

plt.scatter(air_temparature,ac_sales)
plt.title('air_temparature vs. ac_sales')
plt.xlabel('air_temparature')
plt.ylabel('ac_sales')

plt.show()
```
